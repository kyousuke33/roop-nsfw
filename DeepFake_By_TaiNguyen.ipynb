```python
#@title ⚙️ **Cài Đặt Roop & Thư Viện**
# @markdown # 💡 **Tùy chọn phiên bản cài đặt**
# @markdown ---
# @markdown **🔹 Chọn phiên bản cài đặt phù hợp với môi trường của bạn:**
# @markdown - **Cài đặt FULL (không chọn)**: Cài đặt phiên bản đầy đủ.
# @markdown - **Cài Đặt Headless (✔)**: Cài đặt phiên bản nhẹ hơn dành cho server/Colab.

# 🔹 Tùy chọn cài đặt phiên bản headless
Cai_dat_Headless = True  #@param {type:"boolean"}

import os
import time
import subprocess
import sys

# 1️⃣ **Xác định thư mục repo**
repo_path = "/content/roop-nsfw"

# 2️⃣ **Clone repo từ GitHub nếu chưa có**
if not os.path.exists(repo_path):
    print("🔄 Đang clone repository từ GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("✅ Repository đã được clone thành công!\n")
    else:
        print("❌ Lỗi khi clone repo từ GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("⚠️ Repository đã tồn tại, bỏ qua bước clone.\n")

# 3️⃣ **Di chuyển vào thư mục repo**
print("📂 Đang chuyển vào thư mục repository...")
os.chdir(repo_path)
print(f"✅ Đã vào thư mục: {os.getcwd()}\n")

# 4️⃣ **Chọn file requirements dựa trên checkbox**
requirements_file = "requirements-headless.txt" if Cai_dat_Headless else "requirements.txt"

if not os.path.exists(requirements_file):
    print(f"❌ Lỗi: Không tìm thấy `{requirements_file}`. Dừng chương trình.")
    exit()

print(f"📦 Đang đọc danh sách thư viện từ `{requirements_file}`...\n")

with open(requirements_file, "r") as file:
    packages = [line.strip() for line in file if line.strip() and not line.startswith("#")]

total_packages = len(packages)
print(f"🔍 Tổng số thư viện cần cài đặt: {total_packages}\n")

# 5️⃣ **Cài đặt từng thư viện và hiển thị tiến trình**
def print_progress_bar(iteration, total, length=30):
    percent = f"{(iteration / total) * 100:.1f}"
    bar = "█" * int(length * iteration // total) + '-' * (length - int(length * iteration // total))
    sys.stdout.write(f"\r📦 [{iteration}/{total}] [{bar}] {percent}%")
    sys.stdout.flush()

start_time = time.time()

for index, package in enumerate(packages, start=1):
    print(f"\n📦 [{index}/{total_packages}] Đang cài đặt: {package} ...")

    install_process = subprocess.run(
        ["pip", "install", package],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if install_process.returncode == 0:
        elapsed_time = time.time() - start_time
        avg_time_per_package = elapsed_time / index
        estimated_total_time = avg_time_per_package * total_packages
        remaining_time = estimated_total_time - elapsed_time

        print(f"✅ Cài đặt thành công: {package}")
        print(f"⏳ Ước tính thời gian hoàn tất: {remaining_time:.2f}s")
    else:
        print(f"❌ Lỗi khi cài đặt {package}:")
        print(install_process.stderr)
        exit()

    print_progress_bar(index, total_packages)

print("\n🎉 Tất cả thư viện đã được cài đặt thành công! Bạn có thể chạy bảng bên dưới ngay bây giờ. 🚀")
```

`🟡 Sau khi cài đặt xong | Điền tên hình ảnh khuôn mặt - Tên hình ảnh muốn Deepkfake - Tên hình ảnh Output ► RUN`

```python
#@title ⚡️ **Chạy Roop**
import os
import subprocess
import re
from google.colab import drive

# 1️⃣ **Mount Google Drive**
#@markdown ## 🔗 **Mount Google Drive**
#@markdown - **Bắt buộc** để có thể truy cập và lưu file đầu vào/đầu ra.
#@markdown - Google Colab sẽ yêu cầu cấp quyền để truy cập Drive.
print("🔗 Đang mount Google Drive...")
drive.mount('/content/drive')
print("✅ Google Drive đã được mount thành công!\n")

#@markdown ---
#@markdown ## 📂 **Cấu hình Thư Mục & Tên File**
#@markdown - **RootFolder**: Thư mục chứa tất cả các file cần thiết.
#@markdown - **FaceFileName**: Ảnh khuôn mặt để thay thế. (Thư mục Face)
#@markdown - **FSFileName**: Ảnh / Video đích cần thay khuôn mặt. (Hình ảnh: Thư mục Image | Video: Thư mục Video)
#@markdown - **OutputFileName**: Tên file đầu ra sau khi xử lý. (Thư mục Output)

RootFolder = "/content/drive/MyDrive/nsfw-roop"  #@param {type:'string'}
FaceFileName = "face4a.png"      #@param {type:'string'}
FSFileName = "video23.mp4"      #@param {type:'string'}
OutputFileName = "video23a"      #@param {type:'string'}

# 2️⃣ **Clone repo từ GitHub nếu chưa có**
repo_path = RootFolder

if not os.path.exists(repo_path):
    print("🔄 Đang clone repository từ GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("✅ Repository đã được clone thành công!\n")
    else:
        print("❌ Lỗi khi clone repo từ GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("⚠️ Repository đã tồn tại, bỏ qua bước clone.\n")

# 3️⃣ **Tạo thư mục con nếu chưa có**
FaceFolder = os.path.join(RootFolder, "face")
VideoFolder = os.path.join(RootFolder, "video")
ImageFolder = os.path.join(RootFolder, "image")
OutputFolder = os.path.join(RootFolder, "output")

for folder in [FaceFolder, VideoFolder, ImageFolder, OutputFolder]:
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"📂 Đã tạo thư mục: {folder}")

# Chuyển vào thư mục Roop
%cd {RootFolder}

#@markdown ---
#@markdown ## ⚙️ **Cấu hình Xử Lý**
#@markdown - **Device**: Chọn GPU (`cuda`) nếu đang chạy trên máy có GPU, còn không thì chọn CPU.
#@markdown - **Processor**: Kiểu xử lý khuôn mặt, có thể chọn Face Swapper hoặc Face Enhancer.
#@markdown - **VideoEncoder**: Bộ mã hóa video đầu ra, `libx264` thường là lựa chọn tốt nhất.
#@markdown - **VideoQuality**: Chất lượng video đầu ra (1-51), số nhỏ hơn nghĩa là chất lượng cao hơn.

Device = "cpu"                   #@param ["cuda", "cpu"]
Processor = "face_swapper"       #@param ["face_swapper face_enhancer", "face_swapper", "face_enhancer"]
VideoEncoder = "libx264"         #@param ["libx264", "libx265", "libvpx-vp9"]
VideoQuality = "22"              #@param {type:"string"}

# 4️⃣ **Xác định đường dẫn file**
face_path = os.path.join(FaceFolder, FaceFileName)
image_path = os.path.join(ImageFolder, FSFileName)
video_path = os.path.join(VideoFolder, FSFileName)

# 5️⃣ **Kiểm tra sự tồn tại của các file**
print(f"🔍 Kiểm tra file trong thư mục: {RootFolder}")
print(f"✅ File khuôn mặt tồn tại: {os.path.exists(face_path)}")
print(f"✅ File ảnh tồn tại: {os.path.exists(image_path)}")
print(f"✅ File video tồn tại: {os.path.exists(video_path)}")

# 6️⃣ **Xác định nguồn đầu vào và loại file (video hay ảnh)**
video_exists = os.path.exists(video_path)
image_exists = os.path.exists(image_path)

if not os.path.exists(face_path):
    print(f"❌ Lỗi: Không tìm thấy file khuôn mặt '{face_path}'")
    exit()

if video_exists:
    source_path = video_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.mp4")
    print(f"🎬 Đang sử dụng video: {source_path}")
elif image_exists:
    source_path = image_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.png")
    print(f"🖼️ Đang sử dụng ảnh: {source_path}")
else:
    print(f"❌ Lỗi: Không tìm thấy video '{video_path}' hoặc ảnh '{image_path}'")
    exit()

# 7️⃣ **Nếu là video: Kiểm tra số frame và resume nếu cần**
extra_args = []
if video_exists:
    import cv2
    cap = cv2.VideoCapture(video_path)
    total_frames_video = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"📊 Tổng số frame trong video: {total_frames_video}")
    cap.release()

    resume_frame = 1  # Mặc định chạy từ frame 1
    video_name = os.path.splitext(FSFileName)[0]
    temp_dir = os.path.join(VideoFolder, "temp", video_name)

    if os.path.exists(temp_dir):
        print(f"📂 Đã tìm thấy thư mục temp: {temp_dir}")
        frame_files = sorted([f for f in os.listdir(temp_dir) if f.endswith(".png")])
        if frame_files:
            try:
                last_frame = max([int(re.search(r"\d+", f).group()) for f in frame_files if re.search(r"\d+", f)])
                print(f"⏳ Đã xử lý đến frame: {last_frame}")
                resume_frame = last_frame + 1
            except ValueError:
                print("⚠️ Không thể đọc số frame, chạy lại từ đầu.")
    else:
        print("⚠️ Không tìm thấy thư mục temp. Chạy lại từ đầu.")
    extra_args += ["--resume-frame", str(resume_frame)]

# 8️⃣ **Kiểm tra `run.py`**
if not os.path.exists("run.py"):
    print("❌ Lỗi: Không tìm thấy `run.py`. Hãy kiểm tra lại thư mục.")
    exit()

# 9️⃣ **Cấu hình lệnh chạy**
cmd = [
    "python", "run.py",
    "--execution-provider", Device,
    "-s", face_path,
    "-t", source_path,
    "-o", output_path,
    "--frame-processor", Processor,
    "--execution-threads", "16"
]

if video_exists:
    cmd += ["--keep-frames", "--keep-fps"]

cmd += extra_args

print("🚀 Đang chạy lệnh:", " ".join(cmd))
print("📢 Tiến trình đang diễn ra...\n")

# 🔟 **Chạy tiến trình & hiển thị log**
process = subprocess.Popen(
    cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, bufsize=1
)

for line in process.stdout:
    print(line, end="")

process.wait()

# 🔟 **Kiểm tra kết quả**
if process.returncode == 0:
    if os.path.exists(output_path):
        print(f"\n✅ Hoàn tất! File đầu ra được lưu tại: {output_path}")
    else:
        print("\n❌ Không tìm thấy file đầu ra! Có thể `run.py` đã dừng sớm hoặc gặp lỗi.")
else:
    print("\n❌ Có lỗi xảy ra trong quá trình thực thi.")
```
