```python
#@title ‚öôÔ∏è **C√†i ƒê·∫∑t Roop & Th∆∞ Vi·ªán**
# @markdown # üí° **T√πy ch·ªçn phi√™n b·∫£n c√†i ƒë·∫∑t**
# @markdown ---
# @markdown **üîπ Ch·ªçn phi√™n b·∫£n c√†i ƒë·∫∑t ph√π h·ª£p v·ªõi m√¥i tr∆∞·ªùng c·ªßa b·∫°n:**
# @markdown - **C√†i ƒë·∫∑t FULL (kh√¥ng ch·ªçn)**: C√†i ƒë·∫∑t phi√™n b·∫£n ƒë·∫ßy ƒë·ªß.
# @markdown - **C√†i ƒê·∫∑t Headless (‚úî)**: C√†i ƒë·∫∑t phi√™n b·∫£n nh·∫π h∆°n d√†nh cho server/Colab.

# üîπ T√πy ch·ªçn c√†i ƒë·∫∑t phi√™n b·∫£n headless
Cai_dat_Headless = True  #@param {type:"boolean"}

import os
import time
import subprocess
import sys

# 1Ô∏è‚É£ **X√°c ƒë·ªãnh th∆∞ m·ª•c repo**
repo_path = "/content/roop-nsfw"

# 2Ô∏è‚É£ **Clone repo t·ª´ GitHub n·∫øu ch∆∞a c√≥**
if not os.path.exists(repo_path):
    print("üîÑ ƒêang clone repository t·ª´ GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("‚úÖ Repository ƒë√£ ƒë∆∞·ª£c clone th√†nh c√¥ng!\n")
    else:
        print("‚ùå L·ªói khi clone repo t·ª´ GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("‚ö†Ô∏è Repository ƒë√£ t·ªìn t·∫°i, b·ªè qua b∆∞·ªõc clone.\n")

# 3Ô∏è‚É£ **Di chuy·ªÉn v√†o th∆∞ m·ª•c repo**
print("üìÇ ƒêang chuy·ªÉn v√†o th∆∞ m·ª•c repository...")
os.chdir(repo_path)
print(f"‚úÖ ƒê√£ v√†o th∆∞ m·ª•c: {os.getcwd()}\n")

# 4Ô∏è‚É£ **Ch·ªçn file requirements d·ª±a tr√™n checkbox**
requirements_file = "requirements-headless.txt" if Cai_dat_Headless else "requirements.txt"

if not os.path.exists(requirements_file):
    print(f"‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y `{requirements_file}`. D·ª´ng ch∆∞∆°ng tr√¨nh.")
    exit()

print(f"üì¶ ƒêang ƒë·ªçc danh s√°ch th∆∞ vi·ªán t·ª´ `{requirements_file}`...\n")

with open(requirements_file, "r") as file:
    packages = [line.strip() for line in file if line.strip() and not line.startswith("#")]

total_packages = len(packages)
print(f"üîç T·ªïng s·ªë th∆∞ vi·ªán c·∫ßn c√†i ƒë·∫∑t: {total_packages}\n")

# 5Ô∏è‚É£ **C√†i ƒë·∫∑t t·ª´ng th∆∞ vi·ªán v√† hi·ªÉn th·ªã ti·∫øn tr√¨nh**
def print_progress_bar(iteration, total, length=30):
    percent = f"{(iteration / total) * 100:.1f}"
    bar = "‚ñà" * int(length * iteration // total) + '-' * (length - int(length * iteration // total))
    sys.stdout.write(f"\rüì¶ [{iteration}/{total}] [{bar}] {percent}%")
    sys.stdout.flush()

start_time = time.time()

for index, package in enumerate(packages, start=1):
    print(f"\nüì¶ [{index}/{total_packages}] ƒêang c√†i ƒë·∫∑t: {package} ...")

    install_process = subprocess.run(
        ["pip", "install", package],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if install_process.returncode == 0:
        elapsed_time = time.time() - start_time
        avg_time_per_package = elapsed_time / index
        estimated_total_time = avg_time_per_package * total_packages
        remaining_time = estimated_total_time - elapsed_time

        print(f"‚úÖ C√†i ƒë·∫∑t th√†nh c√¥ng: {package}")
        print(f"‚è≥ ∆Ø·ªõc t√≠nh th·ªùi gian ho√†n t·∫•t: {remaining_time:.2f}s")
    else:
        print(f"‚ùå L·ªói khi c√†i ƒë·∫∑t {package}:")
        print(install_process.stderr)
        exit()

    print_progress_bar(index, total_packages)

print("\nüéâ T·∫•t c·∫£ th∆∞ vi·ªán ƒë√£ ƒë∆∞·ª£c c√†i ƒë·∫∑t th√†nh c√¥ng! B·∫°n c√≥ th·ªÉ ch·∫°y b·∫£ng b√™n d∆∞·ªõi ngay b√¢y gi·ªù. üöÄ")
```

```python
#@title ‚ö°Ô∏è **Ch·∫°y Roop**
import os
import subprocess
import re
from google.colab import drive

# 1Ô∏è‚É£ **Mount Google Drive**
#@markdown ## üîó **Mount Google Drive**
#@markdown - **B·∫Øt bu·ªôc** ƒë·ªÉ c√≥ th·ªÉ truy c·∫≠p v√† l∆∞u file ƒë·∫ßu v√†o/ƒë·∫ßu ra.
#@markdown - Google Colab s·∫Ω y√™u c·∫ßu c·∫•p quy·ªÅn ƒë·ªÉ truy c·∫≠p Drive.
print("üîó ƒêang mount Google Drive...")
drive.mount('/content/drive')
print("‚úÖ Google Drive ƒë√£ ƒë∆∞·ª£c mount th√†nh c√¥ng!\n")

#@markdown ---
#@markdown ## üìÇ **C·∫•u h√¨nh Th∆∞ M·ª•c & T√™n File**
#@markdown - **RootFolder**: Th∆∞ m·ª•c ch·ª©a t·∫•t c·∫£ c√°c file c·∫ßn thi·∫øt.
#@markdown - **FaceFileName**: ·∫¢nh khu√¥n m·∫∑t ƒë·ªÉ thay th·∫ø. (Th∆∞ m·ª•c Face)
#@markdown - **FSFileName**: ·∫¢nh / Video ƒë√≠ch c·∫ßn thay khu√¥n m·∫∑t. (H√¨nh ·∫£nh: Th∆∞ m·ª•c Image | Video: Th∆∞ m·ª•c Video)
#@markdown - **OutputFileName**: T√™n file ƒë·∫ßu ra sau khi x·ª≠ l√Ω. (Th∆∞ m·ª•c Output)

RootFolder = "/content/drive/MyDrive/nsfw-roop"  #@param {type:'string'}
FaceFileName = "face4a.png"      #@param {type:'string'}
FSFileName = "video23.mp4"      #@param {type:'string'}
OutputFileName = "video23a"      #@param {type:'string'}

# 2Ô∏è‚É£ **Clone repo t·ª´ GitHub n·∫øu ch∆∞a c√≥**
repo_path = RootFolder

if not os.path.exists(repo_path):
    print("üîÑ ƒêang clone repository t·ª´ GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("‚úÖ Repository ƒë√£ ƒë∆∞·ª£c clone th√†nh c√¥ng!\n")
    else:
        print("‚ùå L·ªói khi clone repo t·ª´ GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("‚ö†Ô∏è Repository ƒë√£ t·ªìn t·∫°i, b·ªè qua b∆∞·ªõc clone.\n")

# 3Ô∏è‚É£ **T·∫°o th∆∞ m·ª•c con n·∫øu ch∆∞a c√≥**
FaceFolder = os.path.join(RootFolder, "face")
VideoFolder = os.path.join(RootFolder, "video")
ImageFolder = os.path.join(RootFolder, "image")
OutputFolder = os.path.join(RootFolder, "output")

for folder in [FaceFolder, VideoFolder, ImageFolder, OutputFolder]:
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"üìÇ ƒê√£ t·∫°o th∆∞ m·ª•c: {folder}")

# Chuy·ªÉn v√†o th∆∞ m·ª•c Roop
%cd {RootFolder}

#@markdown ---
#@markdown ## ‚öôÔ∏è **C·∫•u h√¨nh X·ª≠ L√Ω**
#@markdown - **Device**: Ch·ªçn GPU (`cuda`) n·∫øu ƒëang ch·∫°y tr√™n m√°y c√≥ GPU, c√≤n kh√¥ng th√¨ ch·ªçn CPU.
#@markdown - **Processor**: Ki·ªÉu x·ª≠ l√Ω khu√¥n m·∫∑t, c√≥ th·ªÉ ch·ªçn Face Swapper ho·∫∑c Face Enhancer.
#@markdown - **VideoEncoder**: B·ªô m√£ h√≥a video ƒë·∫ßu ra, `libx264` th∆∞·ªùng l√† l·ª±a ch·ªçn t·ªët nh·∫•t.
#@markdown - **VideoQuality**: Ch·∫•t l∆∞·ª£ng video ƒë·∫ßu ra (1-51), s·ªë nh·ªè h∆°n nghƒ©a l√† ch·∫•t l∆∞·ª£ng cao h∆°n.

Device = "cpu"                   #@param ["cuda", "cpu"]
Processor = "face_swapper"       #@param ["face_swapper face_enhancer", "face_swapper", "face_enhancer"]
VideoEncoder = "libx264"         #@param ["libx264", "libx265", "libvpx-vp9"]
VideoQuality = "22"              #@param {type:"string"}

# 4Ô∏è‚É£ **X√°c ƒë·ªãnh ƒë∆∞·ªùng d·∫´n file**
face_path = os.path.join(FaceFolder, FaceFileName)
image_path = os.path.join(ImageFolder, FSFileName)
video_path = os.path.join(VideoFolder, FSFileName)

# 5Ô∏è‚É£ **Ki·ªÉm tra s·ª± t·ªìn t·∫°i c·ªßa c√°c file**
print(f"üîç Ki·ªÉm tra file trong th∆∞ m·ª•c: {RootFolder}")
print(f"‚úÖ File khu√¥n m·∫∑t t·ªìn t·∫°i: {os.path.exists(face_path)}")
print(f"‚úÖ File ·∫£nh t·ªìn t·∫°i: {os.path.exists(image_path)}")
print(f"‚úÖ File video t·ªìn t·∫°i: {os.path.exists(video_path)}")

# 6Ô∏è‚É£ **X√°c ƒë·ªãnh ngu·ªìn ƒë·∫ßu v√†o v√† lo·∫°i file (video hay ·∫£nh)**
video_exists = os.path.exists(video_path)
image_exists = os.path.exists(image_path)

if not os.path.exists(face_path):
    print(f"‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y file khu√¥n m·∫∑t '{face_path}'")
    exit()

if video_exists:
    source_path = video_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.mp4")
    print(f"üé¨ ƒêang s·ª≠ d·ª•ng video: {source_path}")
elif image_exists:
    source_path = image_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.png")
    print(f"üñºÔ∏è ƒêang s·ª≠ d·ª•ng ·∫£nh: {source_path}")
else:
    print(f"‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y video '{video_path}' ho·∫∑c ·∫£nh '{image_path}'")
    exit()

# 7Ô∏è‚É£ **N·∫øu l√† video: Ki·ªÉm tra s·ªë frame v√† resume n·∫øu c·∫ßn**
extra_args = []
if video_exists:
    import cv2
    cap = cv2.VideoCapture(video_path)
    total_frames_video = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"üìä T·ªïng s·ªë frame trong video: {total_frames_video}")
    cap.release()

    resume_frame = 1  # M·∫∑c ƒë·ªãnh ch·∫°y t·ª´ frame 1
    video_name = os.path.splitext(FSFileName)[0]
    temp_dir = os.path.join(VideoFolder, "temp", video_name)

    if os.path.exists(temp_dir):
        print(f"üìÇ ƒê√£ t√¨m th·∫•y th∆∞ m·ª•c temp: {temp_dir}")
        frame_files = sorted([f for f in os.listdir(temp_dir) if f.endswith(".png")])
        if frame_files:
            try:
                last_frame = max([int(re.search(r"\d+", f).group()) for f in frame_files if re.search(r"\d+", f)])
                print(f"‚è≥ ƒê√£ x·ª≠ l√Ω ƒë·∫øn frame: {last_frame}")
                resume_frame = last_frame + 1
            except ValueError:
                print("‚ö†Ô∏è Kh√¥ng th·ªÉ ƒë·ªçc s·ªë frame, ch·∫°y l·∫°i t·ª´ ƒë·∫ßu.")
    else:
        print("‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y th∆∞ m·ª•c temp. Ch·∫°y l·∫°i t·ª´ ƒë·∫ßu.")
    extra_args += ["--resume-frame", str(resume_frame)]

# 8Ô∏è‚É£ **Ki·ªÉm tra `run.py`**
if not os.path.exists("run.py"):
    print("‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y `run.py`. H√£y ki·ªÉm tra l·∫°i th∆∞ m·ª•c.")
    exit()

# 9Ô∏è‚É£ **C·∫•u h√¨nh l·ªánh ch·∫°y**
cmd = [
    "python", "run.py",
    "--execution-provider", Device,
    "-s", face_path,
    "-t", source_path,
    "-o", output_path,
    "--frame-processor", Processor,
    "--execution-threads", "16"
]

if video_exists:
    cmd += ["--keep-frames", "--keep-fps"]

cmd += extra_args

print("üöÄ ƒêang ch·∫°y l·ªánh:", " ".join(cmd))
print("üì¢ Ti·∫øn tr√¨nh ƒëang di·ªÖn ra...\n")

# üîü **Ch·∫°y ti·∫øn tr√¨nh & hi·ªÉn th·ªã log**
process = subprocess.Popen(
    cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, bufsize=1
)

for line in process.stdout:
    print(line, end="")

process.wait()

# üîü **Ki·ªÉm tra k·∫øt qu·∫£**
if process.returncode == 0:
    if os.path.exists(output_path):
        print(f"\n‚úÖ Ho√†n t·∫•t! File ƒë·∫ßu ra ƒë∆∞·ª£c l∆∞u t·∫°i: {output_path}")
    else:
        print("\n‚ùå Kh√¥ng t√¨m th·∫•y file ƒë·∫ßu ra! C√≥ th·ªÉ `run.py` ƒë√£ d·ª´ng s·ªõm ho·∫∑c g·∫∑p l·ªói.")
else:
    print("\n‚ùå C√≥ l·ªói x·∫£y ra trong qu√° tr√¨nh th·ª±c thi.")
```
