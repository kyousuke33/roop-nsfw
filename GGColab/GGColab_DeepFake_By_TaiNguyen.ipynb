```python
#@title âš™ï¸ **CÃ i Äáº·t Roop & ThÆ° Viá»‡n**
# @markdown # ğŸ’¡ **TÃ¹y chá»n phiÃªn báº£n cÃ i Ä‘áº·t**
# @markdown ---
# @markdown **ğŸ”¹ Chá»n phiÃªn báº£n cÃ i Ä‘áº·t phÃ¹ há»£p vá»›i mÃ´i trÆ°á»ng cá»§a báº¡n:**
# @markdown - **CÃ i Ä‘áº·t FULL (khÃ´ng chá»n)**: CÃ i Ä‘áº·t phiÃªn báº£n Ä‘áº§y Ä‘á»§.
# @markdown - **CÃ i Äáº·t Headless (âœ”)**: CÃ i Ä‘áº·t phiÃªn báº£n nháº¹ hÆ¡n dÃ nh cho server/Colab.

# ğŸ”¹ TÃ¹y chá»n cÃ i Ä‘áº·t phiÃªn báº£n headless
Cai_dat_Headless = False  #@param {type:"boolean"}
import os
import time
import subprocess
import sys
repo_path = "/content/roop-nsfw"
if not os.path.exists(repo_path):
    print("ğŸ”„ Äang clone repository tá»« GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("âœ… Repository Ä‘Ã£ Ä‘Æ°á»£c clone thÃ nh cÃ´ng!\n")
    else:
        print("âŒ Lá»—i khi clone repo tá»« GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("âš ï¸ Repository Ä‘Ã£ tá»“n táº¡i, bá» qua bÆ°á»›c clone.\n")
print("ğŸ“‚ Äang chuyá»ƒn vÃ o thÆ° má»¥c repository...")
os.chdir(repo_path)
print(f"âœ… ÄÃ£ vÃ o thÆ° má»¥c: {os.getcwd()}\n")
requirements_file = "requirements-headless.txt" if Cai_dat_Headless else "requirements.txt"

if not os.path.exists(requirements_file):
    print(f"âŒ Lá»—i: KhÃ´ng tÃ¬m tháº¥y `{requirements_file}`. Dá»«ng chÆ°Æ¡ng trÃ¬nh.")
    exit()

print(f"ğŸ“¦ Äang Ä‘á»c danh sÃ¡ch thÆ° viá»‡n tá»« `{requirements_file}`...\n")

with open(requirements_file, "r") as file:
    packages = [line.strip() for line in file if line.strip() and not line.startswith("#")]

total_packages = len(packages)
print(f"ğŸ” Tá»•ng sá»‘ thÆ° viá»‡n cáº§n cÃ i Ä‘áº·t: {total_packages}\n")

def print_progress_bar(iteration, total, length=30):
    percent = f"{(iteration / total) * 100:.1f}"
    bar = "â–ˆ" * int(length * iteration // total) + '-' * (length - int(length * iteration // total))
    sys.stdout.write(f"\rğŸ“¦ [{iteration}/{total}] [{bar}] {percent}%")
    sys.stdout.flush()

start_time = time.time()

for index, package in enumerate(packages, start=1):
    print(f"\nğŸ“¦ [{index}/{total_packages}] Äang cÃ i Ä‘áº·t: {package} ...")

    install_process = subprocess.run(
        ["pip", "install", package],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if install_process.returncode == 0:
        elapsed_time = time.time() - start_time
        avg_time_per_package = elapsed_time / index
        estimated_total_time = avg_time_per_package * total_packages
        remaining_time = estimated_total_time - elapsed_time

        print(f"âœ… CÃ i Ä‘áº·t thÃ nh cÃ´ng: {package}")
        print(f"â³ Æ¯á»›c tÃ­nh thá»i gian hoÃ n táº¥t: {remaining_time:.2f}s")
    else:
        print(f"âŒ Lá»—i khi cÃ i Ä‘áº·t {package}:")
        print(install_process.stderr)
        exit()

    print_progress_bar(index, total_packages)

print("\nğŸ‰ Táº¥t cáº£ thÆ° viá»‡n Ä‘Ã£ Ä‘Æ°á»£c cÃ i Ä‘áº·t thÃ nh cÃ´ng! Báº¡n cÃ³ thá»ƒ cháº¡y báº£ng bÃªn dÆ°á»›i ngay bÃ¢y giá». ğŸš€")
```

```python
#@title âš¡ï¸ **Cháº¡y Roop**
import os
import subprocess
import re
from google.colab import drive

# 1ï¸âƒ£ **Mount Google Drive**
#@markdown ## ğŸ”— **Mount Google Drive**
#@markdown - **Báº¯t buá»™c** Ä‘á»ƒ cÃ³ thá»ƒ truy cáº­p vÃ  lÆ°u file Ä‘áº§u vÃ o/Ä‘áº§u ra.
#@markdown - Google Colab sáº½ yÃªu cáº§u cáº¥p quyá»n Ä‘á»ƒ truy cáº­p Drive.
print("ğŸ”— Äang mount Google Drive...")
drive.mount('/content/drive')
print("âœ… Google Drive Ä‘Ã£ Ä‘Æ°á»£c mount thÃ nh cÃ´ng!\n")

#@markdown ---
#@markdown ## ğŸ“‚ **Cáº¥u hÃ¬nh ThÆ° Má»¥c & TÃªn File**
#@markdown - **RootFolder**: ThÆ° má»¥c chá»©a táº¥t cáº£ cÃ¡c file cáº§n thiáº¿t.
#@markdown - **FaceFileName**: áº¢nh khuÃ´n máº·t Ä‘á»ƒ thay tháº¿. Pháº£i cÃ³ Ä‘uÃ´i file (.jpg,.png) (ThÆ° má»¥c Face)
#@markdown - **FSFileName**: áº¢nh / Video Ä‘Ã­ch cáº§n thay khuÃ´n máº·t. Pháº£i cÃ³ Ä‘uÃ´i file Ä‘á»ƒ phÃ¢n biá»‡t hÃ¬nh áº£nh (.jpg,.png) vÃ  video (.mp4)  (HÃ¬nh áº£nh: ThÆ° má»¥c Image | Video: ThÆ° má»¥c Video)
#@markdown - **OutputFileName**: TÃªn file Ä‘áº§u ra sau khi xá»­ lÃ½. KhÃ´ng cáº§n Ä‘uÃ´i file (ThÆ° má»¥c Output)

RootFolder = "/content/drive/MyDrive/nsfw-roop"  #@param {type:'string'}
FaceFileName = "face1a.png"      #@param {type:'string'}
FSFileName = "image1.png"      #@param {type:'string'}
OutputFileName = "image1"      #@param {type:'string'}

repo_path = RootFolder
if not os.path.exists(repo_path):
    print("ğŸ”„ Äang clone repository tá»« GitHub...")
    clone_process = subprocess.run(
        ["git", "clone", "https://github.com/kyousuke33/roop-nsfw.git", repo_path],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    if clone_process.returncode == 0:
        print("âœ… Repository Ä‘Ã£ Ä‘Æ°á»£c clone thÃ nh cÃ´ng!\n")
    else:
        print("âŒ Lá»—i khi clone repo tá»« GitHub:")
        print(clone_process.stderr)
        exit()
else:
    print("âš ï¸ Repository Ä‘Ã£ tá»“n táº¡i, bá» qua bÆ°á»›c clone.\n")

%cd {RootFolder}

FaceFolder = os.path.join(RootFolder, "face")
VideoFolder = os.path.join(RootFolder, "video")
ImageFolder = os.path.join(RootFolder, "image")
OutputFolder = os.path.join(RootFolder, "output")
for folder in [FaceFolder, VideoFolder, ImageFolder, OutputFolder]:
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"ğŸ“‚ ÄÃ£ táº¡o thÆ° má»¥c: {folder}")


#@markdown ---
#@markdown ## âš™ï¸ **Cáº¥u hÃ¬nh Xá»­ LÃ½**
#@markdown - **Device**: Chá»n GPU (`cuda`) náº¿u Ä‘ang cháº¡y trÃªn mÃ¡y cÃ³ GPU, cÃ²n khÃ´ng thÃ¬ chá»n CPU.
#@markdown - **Processor**: Kiá»ƒu xá»­ lÃ½ khuÃ´n máº·t, cÃ³ thá»ƒ chá»n Face Swapper hoáº·c Face Enhancer.
#@markdown - **VideoEncoder**: Bá»™ mÃ£ hÃ³a video Ä‘áº§u ra, `libx264` thÆ°á»ng lÃ  lá»±a chá»n tá»‘t nháº¥t.
#@markdown - **VideoQuality**: Cháº¥t lÆ°á»£ng video Ä‘áº§u ra (1-51), sá»‘ nhá» hÆ¡n nghÄ©a lÃ  cháº¥t lÆ°á»£ng cao hÆ¡n.

Device = "cpu"                   #@param ["cuda", "cpu"]
Processor = "face_swapper"       #@param ["face_swapper face_enhancer", "face_swapper", "face_enhancer"]
VideoEncoder = "libx264"         #@param ["libx264", "libx265", "libvpx-vp9"]
VideoQuality = "22"              #@param {type:"string"}

face_path = os.path.join(FaceFolder, FaceFileName)
image_path = os.path.join(ImageFolder, FSFileName)
video_path = os.path.join(VideoFolder, FSFileName)
print(f"ğŸ” Kiá»ƒm tra file trong thÆ° má»¥c: {RootFolder}")
print(f"âœ… File khuÃ´n máº·t tá»“n táº¡i: {os.path.exists(face_path)}")
print(f"âœ… File áº£nh tá»“n táº¡i: {os.path.exists(image_path)}")
print(f"âœ… File video tá»“n táº¡i: {os.path.exists(video_path)}")
video_exists = os.path.exists(video_path)
image_exists = os.path.exists(image_path)

if not os.path.exists(face_path):
    print(f"âŒ Lá»—i: KhÃ´ng tÃ¬m tháº¥y file khuÃ´n máº·t '{face_path}'")
    exit()

if video_exists:
    source_path = video_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.mp4")
    print(f"ğŸ¬ Äang sá»­ dá»¥ng video: {source_path}")
elif image_exists:
    source_path = image_path
    output_path = os.path.join(OutputFolder, f"{OutputFileName}-fs.png")
    print(f"ğŸ–¼ï¸ Äang sá»­ dá»¥ng áº£nh: {source_path}")
else:
    print(f"âŒ Lá»—i: KhÃ´ng tÃ¬m tháº¥y video '{video_path}' hoáº·c áº£nh '{image_path}'")
    exit()

extra_args = []
if video_exists:
    import cv2
    cap = cv2.VideoCapture(video_path)
    total_frames_video = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"ğŸ“Š Tá»•ng sá»‘ frame trong video: {total_frames_video}")
    cap.release()

    resume_frame = 1
    video_name = os.path.splitext(FSFileName)[0]
    temp_dir = os.path.join(VideoFolder, "temp", video_name)

    if os.path.exists(temp_dir):
        print(f"ğŸ“‚ ÄÃ£ tÃ¬m tháº¥y thÆ° má»¥c temp: {temp_dir}")
        frame_files = sorted([f for f in os.listdir(temp_dir) if f.endswith(".png")])
        if frame_files:
            try:
                last_frame = max([int(re.search(r"\d+", f).group()) for f in frame_files if re.search(r"\d+", f)])
                print(f"â³ ÄÃ£ xá»­ lÃ½ Ä‘áº¿n frame: {last_frame}")
                resume_frame = last_frame + 1
            except ValueError:
                print("âš ï¸ KhÃ´ng thá»ƒ Ä‘á»c sá»‘ frame, cháº¡y láº¡i tá»« Ä‘áº§u.")
    else:
        print("âš ï¸ KhÃ´ng tÃ¬m tháº¥y thÆ° má»¥c temp. Cháº¡y láº¡i tá»« Ä‘áº§u.")
    extra_args += ["--resume-frame", str(resume_frame)]

if not os.path.exists("run.py"):
    print("âŒ Lá»—i: KhÃ´ng tÃ¬m tháº¥y `run.py`. HÃ£y kiá»ƒm tra láº¡i thÆ° má»¥c.")
    exit()

cmd = [
    "python", "run.py",
    "--execution-provider", Device,
    "-s", face_path,
    "-t", source_path,
    "-o", output_path,
    "--frame-processor", Processor,
    "--execution-threads", "16"
]

if video_exists:
    cmd += ["--keep-frames", "--keep-fps"]

cmd += extra_args

print("ğŸš€ Äang cháº¡y lá»‡nh:", " ".join(cmd))
print("ğŸ“¢ Tiáº¿n trÃ¬nh Ä‘ang diá»…n ra...\n")

process = subprocess.Popen(
    cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, bufsize=1
)

for line in process.stdout:
    print(line, end="")

process.wait()

if process.returncode == 0:
    if os.path.exists(output_path):
        print(f"\nâœ… HoÃ n táº¥t! File Ä‘áº§u ra Ä‘Æ°á»£c lÆ°u táº¡i: {output_path}")
    else:
        print("\nâŒ KhÃ´ng tÃ¬m tháº¥y file Ä‘áº§u ra! CÃ³ thá»ƒ `run.py` Ä‘Ã£ dá»«ng sá»›m hoáº·c gáº·p lá»—i.")
else:
    print("\nâŒ CÃ³ lá»—i xáº£y ra trong quÃ¡ trÃ¬nh thá»±c thi.")
```
